```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
newdata = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Embarked','Survived'], axis = 1)
```

```python
newdata['Cabin'].fillna('No', inplace=True)
newdata.head()
```

```python
target = df['Survived']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Cabin = LabelEncoder()
le_Sex = LabelEncoder()
```

```python
newdata['Cabin_n'] = le_Cabin.fit_transform(newdata['Cabin'])
newdata['Sex_n'] = le_Sex.fit_transform(newdata['Sex'])
newdata
```

```python
traindata = newdata.drop(['Sex','Cabin'], axis = 1)
traindata['Age'].fillna(0, inplace=True)
traindata
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(traindata,target)
```

```python
model.score(traindata,target)
```

```python
model.predict([[1,25,71.2833,81,1]])
```
